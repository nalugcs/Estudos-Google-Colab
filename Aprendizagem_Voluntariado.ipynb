    ctype, encoding = mimetypes.guess_type("Relatório Snackin.pdf")

    maintype, subtype = ctype.split('/', 1)
    if maintype == 'text':
        with open("Relatório.pdf") as f:
            mime = MIMEText(f.read(), _subtype=subtype)
    elif maintype == 'image':
        with open("Relatório.pdf", 'rb') as f:
            mime = MIMEImage(f.read(), _subtype=subtype)
    elif maintype == 'audio':
        with open("Relatório.pdf", 'rb') as f:
            mime = MIMEAudio(f.read(), _subtype=subtype)
    else:
        with open("Relatório.pdf", 'rb') as f:
            mime = MIMEBase(maintype, subtype)
            mime.set_payload(f.read())
        encoders.encode_base64(mime)
    mime.add_header('Content-Disposition', 'attachment', filename="Relatório Snackin.pdf")
    msg.attach(mime)
